##### Further reserach
To find the words used most often in the text 


```python
import collections
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

file = open('horror.txt', encoding="utf8")
a= file.read()

stopwords = set(line.strip() for line in open('stopwords.txt'))

wordcount = {}
for word in a.lower().split():
    word = word.replace(".","")
    word = word.replace(",","")
    word = word.replace(":","")
    word = word.replace("\"","")
    word = word.replace("!","")
    word = word.replace("","")
    word = word.replace("","")
    word = word.replace("*","")
    if word not in stopwords:
        if word not in wordcount:
            wordcount[word] = 1
        else:
            wordcount[word] += 1

n_print = int(10)
print("\nOK. The {} most common words are as follows\n".format(n_print))
word_counter = collections.Counter(wordcount)
for word, count in word_counter.most_common(n_print):
    print(word, ": ", count)

file.close()

lst = word_counter.most_common(n_print)
df = pd.DataFrame(lst, columns = ['Word', 'Count'])
df.plot.bar(x='Word',y='Count')
```

###### Conclusion of further research 
It is clear from the data which was gathered that the most frequent words in the chosen text, are not in facts words which necessarily evoke in the reader, the feeling of fear, dread or uneasiness. Many of the words are nouns such as *house*, *hand*, *room* etc. which in the contect of a horror setting may evoke the eerie feeling we  are expecting in a horror text. However, on their own, they have no such connotations. However, in this case, the word appearing most frequently in the text was the word *ghost*, which connects back to the title of the text.

##### PoS Tagging 
In my analysis I also wanted to complete some PoS tagging of the first three sentences, of the first short story in the collection, which was *The Black Cat* by Edgar Alan Poe.

```python
import nltk
from nltk.tokenize import word_tokenize
black_cat = """""For the most wild, yet most homely narrative which I am about to pen, I
neither expect nor solicit belief. Mad indeed would I be to expect it,
in a case where my very senses reject their own evidence. Yet, mad am I
notand very surely do I not dream. """""
words_in_black_cat = word_tokenize(black_cat)
```

```python
import nltk
nltk.pos_tag(words_in_black_cat)
```

From this we can conclude that the most frequent parts of speech used in this part of the text where adverbs/superlatives- **7**, adjectives/numerals- **7**, pronoun/personal pronoun- **8** and verb base form and verb present tense- **9**.


We can conclude, that there are many things we can observe and explore using the BeautifulSoup,NLTK and Regex packages. In regards to horror stories,one can see that it is not necessarily the *scary words*, which build the atmosphere of a horror story, it may be the case of other aspect which together create such atmosphere. From the small text sample which we analysed with PoS, it is clear that the most frequent parts of speech used are verbs- and in turn adverbs. If we were to generalize and assume this would be the case for the whole text, the implications would be that the *essence* of what makes a horror story *scary* is its usage and description of actions.
